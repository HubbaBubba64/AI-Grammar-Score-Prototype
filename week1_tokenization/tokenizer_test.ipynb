```python
from transformers import AutoTokenizer 

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text = "This is a simple sentence for testing."

tokens = tokenizer(text)
print("Token IDs:", tokens['input_ids'])
print("Tokens:", tokenizer.convert_ids_to_tokens(tokens['input_ids']))
```

```python
bad_sentence = "This a simple sentence test."

bad_tokens = tokenizer(bad_sentence)

print("Token Bad:", tokenizer.convert_ids_to_tokens(bad_tokens['input_ids']))

```

## Note

- CLS token shows the entire sentence for BERT.
- Texts must be converted into tokens and IDS so the model can understand it.
- Small changes can produce different token sequences.
