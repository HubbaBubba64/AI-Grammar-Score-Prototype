```python
import torch
from transformers import AutoTokenizer, AutoModel
from torch.nn.functional import cosine_similarity

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")


text_good = "This is a simple sentence for testing."
text_bad = "This a simple sentence test."

inputs_good = tokenizer(text_good, return_tensors="pt")
inputs_bad = tokenizer(text_bad, return_tensors="pt")

with torch.no_grad():
    outputs_good = model(**inputs_good)
    outputs_bad = model(**inputs_bad)

cls_good = outputs_good.last_hidden_state[:, 0, :]
cls_bad = outputs_bad.last_hidden_state[:, 0, :]
similarity = cosine_similarity(cls_good, cls_bad)

print("CLS good shape:", cls_good.shape)
print("CLS bad shape:", cls_bad.shape)
print("Similarity Score:", similarity.item())
```
